```python
import pandas as pd
import yfinance as yf
```

```python
prices = yf.download("^GSPC ^NDX", start="1985-10-01", end="2022-05-03")['Adj Close']
prices.columns = ['SPX', 'NDX']
```

```python
returns = prices.pct_change()
returns.dropna(inplace=True)
returns
```

```python
#resample daily data to calulate weekly returns
wkly_returns = pd.DataFrame(returns.resample("W").apply(lambda x: ((x + 1).cumprod() - 1).last("D")))
wkly_returns.head()
```

```python
counter = 0
def downWeeks(x):
    global counter
    if x >= 0:
        counter = 0
        return counter
    else:
        counter += 1
        return counter
        
```

```python
wkly_returns['SPX down weeks'] = wkly_returns['SPX'].apply(downWeeks)
wkly_returns['NDX down weeks'] = wkly_returns['NDX'].apply(downWeeks)
wkly_returns
```

```python
wkly_returns.loc[(wkly_returns['SPX down weeks'] >=5) & (wkly_returns['NDX down weeks'] >= 5),['SPX down weeks', 'NDX down weeks']]
```
