```python
import yfinance as yf
import pandas as pd
import numpy as np


TICKER = "^GSPC"
```

```python
ticker = yf.Ticker(TICKER).history(period="max")
ticker = ticker.loc[:, ["Close"]]
ticker.columns = ['close']
ticker.index.name = 'date'
ticker['return'] = ticker['close'].pct_change()
ticker['upDn'] = np.where(ticker['return'] >= 0, "Up", "Down")
ticker['tradingDay'] = ticker.index.day_of_week.astype('str') + ' ' + ticker.index.day_name()
ticker['nextDayReturn'] = ticker['close'].pct_change().shift(-1)
ticker['nextDayUpDn'] = np.where(ticker['nextDayReturn'] >= 0, "Up", "Down")

ticker
```

```python
ticker.loc[ticker.index >= '2020',['tradingDay', 'upDn','nextDayReturn', 'nextDayUpDn']].groupby(['tradingDay','upDn','nextDayUpDn']).agg(count = ('nextDayReturn', 'count'),median_return = ('nextDayReturn', 'median'))
```

```python
ticker.loc[ticker.index >= '2020',['tradingDay', 'return']].groupby('tradingDay').agg(count = ('return', 'count'),median_return = ('return', 'median'))
```
